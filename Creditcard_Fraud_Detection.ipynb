Importing the dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loading the dataset into pandas dataframe
credit_card_dataset = pd.read_csv('/content/creditcard.csv')
```

```python
#first five rows of dataframe
credit_card_dataset.head()
```

```python
#last five rows of dataframe
credit_card_dataset.tail()
```

```python
#dataset information
credit_card_dataset.info()
```

```python
#checking the numbers of missing values
credit_card_dataset.isnull().sum()
```

```python
#checking the distribution of legit transaction and fraudulent transaction
credit_card_dataset['Class'].value_counts()
```

This dataset is highly unbalance

0 --> normal transaction
1 --> fraudulent transaction

seperating the normal and fraudulent data from column 'Class'

```python
legit = credit_card_dataset[credit_card_dataset.Class == 0]
fraud = credit_card_dataset[credit_card_dataset.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#statistical measures of data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

we saw here that mean amount has a huge difference so we further process it

```python
#comparing the values for both transaction
credit_card_dataset.groupby('Class').mean()
```

Under-sampling (for unbalanced dataset)

Build a sample dataset containing similar distribution of a normal(legit) transactions and fraudulent transaction

No. of fraudulent transaction ---> 492

```python
legit_sample = legit.sample(n=492)
```

Concatenating two dataframe

```python
new_dataset = pd.concat([legit_sample, fraud], axis = 0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Splitting the data into features and target

```python
X = new_dataset.drop(['Class'], axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into training data and test data

```python
X_train, X_test, Y_train,Y_test = train_test_split(X,Y, test_size = 0.2, stratify=Y, random_state=2)
```

```python
print(X_train.shape, X_test.shape)
```

Model training

Logistic regression

```python
model = LogisticRegression()
```

```python
#training the Logistic Regression with training data
model.fit(X_train , Y_train)
```

Model Evaluation

Accuracy score

```python
#Accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)

```

```python
print('Accuracy on training data ',training_data_accuracy )
```

```python
#Accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)

```

```python
print('Accuracy on training data ',test_data_accuracy )
```
